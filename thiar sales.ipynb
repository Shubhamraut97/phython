# exploratory data analysis

# ##gender

```python
cdf.columns
```

```python
ax=sns.countplot(x='Gender',data=cdf)
```

```python
ax=sns.countplot(x='Gender',data=cdf)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=cdf.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen) 
```

```python
cdf.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

# age


```python
ax=sns.countplot(x='Age Group',hue='Gender',data=cdf)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=cdf.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

# state

```python
sales_state=cdf.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_order=cdf.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(x='State',y='Amount',data=sales_order)
```

# marital status

```python
ax=sns.countplot(x='Marital_Status',data=cdf)
sns.set(rc={'figure.figsize':(15,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=cdf.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_age)
```

```python
ax=sns.countplot(x='Occupation',data=cdf)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_order=cdf.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(x='Occupation',y='Amount',data=sales_order)
```

```python
ax=sns.countplot(x='Product_Category',data=cdf)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_order=cdf.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(x='Product_Category',y='Amount',data=sales_order)
```

```python
sales_id=cdf.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(x='Product_ID',y='Orders',data=sales_id)
```
